```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
import glob
glob.glob("pv_images/*")
```

```python
results = model.predict(
    source="pv_images",
    imgsz=640,
    conf=0.25,
    save=True
)
```

```python
import glob
import matplotlib.pyplot as plt
import os

out_dir = "runs/detect/predict"
imgs = sorted(glob.glob(out_dir + "/*.jpg")) + sorted(glob.glob(out_dir + "/*.png"))

for p in imgs:
    img = plt.imread(p)
    plt.figure(figsize=(8, 4))
    plt.imshow(img)
    plt.axis("off")
    plt.title(os.path.basename(p))
    plt.show()
```
